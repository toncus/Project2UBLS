```python
#orient : str {'dict', 'list', 'series', 'split', 'records', 'index'}
records = df[df['OCC_GROUP']=='major'][['OCC-Group',"OCC_TITLE"]].drop_duplicates().set_index('OCC-Group')
major_group = records["OCC_TITLE"].to_dict()
```

```python
def return_OCC_Major(number):
    try:
        return major_group[number]
    except:
        return ""
```

```python
df2 = df
df2 = df2[df2["TOT_EMP"]!="**"]
df2['Major Group'] = df2['OCC-Group'].map(return_OCC_Major)
df2["TOT_EMP"]=df2["TOT_EMP"].map(int)
df2.head()
```

```python
df3 = df2.groupby(["STATE","Major Group","OCC_TITLE"]).mean()[["TOT_EMP"]]
```

```python
stateJobs2016 = []
for state in states:
    majorjobs = []
    for mj in [item for item in major_group.values()]:
        minorjobs = []
        minor_group = df4.loc[state,mj,:].reset_index()["OCC_TITLE"]
        for job in minor_group:
            minorjobs.append({"name":job,"value":str(df4.loc[state,mj,job]["TOT_EMP"])})
        majorjobs.append({"name":mj,"children":minorjobs})
    stateJobs2016.append({"name":state,"children":majorjobs})
```
